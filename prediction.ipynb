```python
import numpy as np
import time
```

```python
# Load Model Back To Test

scaler = joblib.load("scaler_used_for_simulation.pkl")  
loaded_model = joblib.load("rain_intensity_model.pkl")
```

```python

intensity_labels = {
    0: "No Rain",
    1: "Light Rain",
    2: "Heavy Rain"
}

print("\nREAL-TIME RAIN INTENSITY PREDICTION (SIMULATION MODE)\n")

# Generate 10 random simulated inputs for demo
for i in range(10):
    # Simulated sensor values (similar range as your training data)
    sensor_value = np.random.uniform(0, 1, size=(1, 12))

    # Scale input using training scaler
    sensor_scaled = scaler.transform(sensor_value)

    # Predict using your already loaded model
    pred_numeric = loaded_model.predict(sensor_scaled)[0]
    pred_label = intensity_labels[pred_numeric]

    print(f"Sample {i+1}:")
    print(f"  Sensor Value      : {sensor_value[0][0]:.4f}")
    print(f"  Predicted Class   : {pred_numeric}")
    print(f"  Rain Intensity    : {pred_label}")
    print("-" * 50)

    time.sleep(1)  # small delay for effect
```
